```python
from surprise import KNNBasic, Dataset, Reader, accuracy,KNNWithMeans
from surprise.model_selection import train_test_split
import pandas as pd
import time
# Load the data
ratings = pd.read_csv('/kaggle/input/the-movies-dataset/ratings_small.csv')  # Please replace with your actual path
reader = Reader(rating_scale=(ratings.rating.min(), ratings.rating.max()))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
# Define the algorithm
sim_options = {
    'user_based': False  # Compute similarities between users
}
algo = KNNWithMeans(sim_options=sim_options)

# Create a train-test split
trainset, testset = train_test_split(data, test_size=0.2,random_state = 42)
start_total_time = time.time()

# Train the algorithm on the trainset
algo.fit(trainset)

# Predict ratings for the testset
predictions = algo.test(testset)
end_total_time = time.time()
# Then compute RMSE and MAE
print('RMSE: ', accuracy.rmse(predictions))
print('MAE: ', accuracy.mae(predictions))
print(f"Total time taken: {end_total_time - start_total_time:.2f} seconds.")
```
